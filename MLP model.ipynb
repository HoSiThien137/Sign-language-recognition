```python
import pandas as pd
from enum import Enum
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense, Dropout, BatchNormalization, LSTM
from keras.optimizers import Adam
from keras.callbacks import Callback
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, f1_score, precision_score, recall_score

class Actions(Enum):
    Opaque = 0 
    Red = 1 
    Green = 2 
    Yellow = 3 
    Bright = 4 
    Light_blue = 5 
    Colors = 6 
    Red2 = 7
    Women = 8 
    Enemy = 9 
    Son = 10 
    Man = 11 
    Away = 12 
    Drawer = 13 
    Born = 14 
    Learn = 15 
    Call = 16 
    Skimmer = 17 
    Bitter = 18 
    Sweet_milk = 19 
    Milk = 20 
    Water = 21 
    Food = 22 
    Argentina = 23 
    Urugauy = 24
    Country = 25 
    Last_name = 26 
    Where = 27 
    Mock = 28 
    Birthday = 29 
    Breakfast = 30 
    Photo = 31 
    Hungry = 32 
    Map = 33 
    Coin = 34 
    Music = 35 
    Ship = 36 
    none = 37 
    Name = 38 
    Patience = 39 
    Perfume = 40 
    Deaf = 41 
    Trap = 42 
    Rice = 43 
    Barbecue = 44 
    Candy = 45 
    Chewing_gum = 46 
    Spaghetti = 47 
    Yogurt = 48 
    Accept = 49 
    Thanks = 50 
    Shut_down = 51 
    Appear = 52 
    To_land = 53 
    Catch = 54 
    Help = 55 
    Dance = 56 
    Bathe = 57 
    Buy = 58 
    Copy = 59 
    Run = 60 
    Realize = 61 
    Give = 62 
    Find = 63 

class LossHistory(Callback):
    def on_train_begin(self, logs={}):
        self.losses = {'batch':[], 'epoch':[]}
        self.accuracy = {'batch':[], 'epoch':[]}
        self.val_loss = {'batch':[], 'epoch':[]}
        self.val_acc = {'batch':[], 'epoch':[]}

    def on_batch_end(self, batch, logs={}):
        self.losses['batch'].append(logs.get('loss'))
        self.accuracy['batch'].append(logs.get('acc'))
        self.val_loss['batch'].append(logs.get('val_loss'))
        self.val_acc['batch'].append(logs.get('val_acc'))

    def on_epoch_end(self, batch, logs={}):
        self.losses['epoch'].append(logs.get('loss'))
        self.accuracy['epoch'].append(logs.get('acc'))
        self.val_loss['epoch'].append(logs.get('val_loss'))
        self.val_acc['epoch'].append(logs.get('val_acc'))

    def loss_plot(self, loss_type):
        plt.figure(figsize=(12, 8))  # Điều chỉnh kích thước của hình ảnh
        plt.style.use('ggplot')  # Sử dụng ggplot style với nền màu trắng
        iters = range(len(self.losses[loss_type]))
        plt.plot(iters, self.accuracy[loss_type], 'r', label='train acc')
        plt.plot(iters, self.losses[loss_type], 'g', label='train loss')
        if loss_type == 'epoch':
            plt.plot(iters, self.val_acc[loss_type], 'b', label='val acc')
            plt.plot(iters, self.val_loss[loss_type], 'k', label='val loss')
        plt.grid(True)
        plt.xlabel(loss_type, fontsize=14)  # Điều chỉnh kích thước của nhãn trục x
        plt.ylabel('acc-loss', fontsize=14)  # Điều chỉnh kích thước của nhãn trục y
        plt.legend(loc="upper right", fontsize=12)
        plt.savefig('/kaggle/working/loss_accuracy_plot.png', bbox_inches="tight", transparent=True)

def get_report(y_true, y_pred, classes):
    clf_report = classification_report(y_true, y_pred, labels=classes, zero_division=0)
    clf_report = clf_report.replace('\n\n', '\n')
    clf_report = clf_report.replace('micro avg', 'micro_avg')
    clf_report = clf_report.replace('macro avg', 'macro_avg')
    clf_report = clf_report.replace('weighted avg', 'weighted_avg')
    clf_report = clf_report.replace(' / ', '/')
    lines = clf_report.split('\n')

    class_names, plotMat, support = [], [], []
    for line in lines[1:]:
        t = line.strip().split()
        if len(t) < 2:
            continue
        v = [float(x) for x in t[1: len(t) - 1]]
        if len(v) == 1 : v = v * 3
        support.append(int(t[-1]))
        class_names.append(t[0])
        plotMat.append(v)
    plotMat = np.array(plotMat)
    support = np.array(support)
    return class_names, plotMat, support

def plot_confusion_matrix(classes, mat, normalize=True, cmap=plt.cm.Blues):
    cm = np.copy(mat)
    title = 'Ma trận nhầm lẫn (không chuẩn hóa)'
    if normalize: 
        cm = cm.astype('float') / np.sum(cm, axis=1, keepdims=True)
        title = title.replace('không', 'có')
    plt.clf()
    fig, ax = plt.subplots(figsize=(30, 25))  # Điều chỉnh kích thước hình ảnh
    ax.set_title(title, y=-0.06, fontsize=22)
    ax.xaxis.set_ticks_position('top')
    ax.xaxis.set_label_position('top')
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.clim(vmin=0.0, vmax=1.0)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=90, fontsize=12)  # Điều chỉnh góc và kích thước của nhãn
    plt.yticks(tick_marks, classes, fontsize=12)
    fmt = '.2f' if normalize else 'd'
    thresh = np.max(cm) / 2
    thresh = 1 / 2
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            color = "white" if (cm[i, j] > thresh) else "black"
            plt.text(j, i, format(cm[i, j], fmt), horizontalalignment="center", color=color, fontsize=10)  # Điều chỉnh kích thước của con số
    plt.ylabel('Nhãn thực', fontsize=16)  # Điều chỉnh kích thước của nhãn trục y
    plt.xlabel('Nhãn dự đoán', fontsize=16)  # Điều chỉnh kích thước của nhãn trục x
    plt.tight_layout()
    plt.savefig('/kaggle/working/confusion_matrix.png', bbox_inches="tight", transparent=True)

def get_scores(y_true, y_pred, classes):
    correct, wrong = {}, {}
    for tag in classes:
        correct[tag] = 0
        wrong[tag] = 0

    for tag, pred in zip(y_true, y_pred):
        if tag == pred:
            correct[tag] += 1
        else:
            wrong[tag] += 1

    scores = []
    total = len(y_true)
    for tag in classes:
        cur = np.array([correct[tag], wrong[tag]])
        scores.append(cur / total)
    return np.array(scores)

def plot_tag_scores(classes, scores, normalize=True):
    plt.clf()
    width = 0.45
    fig, ax = plt.subplots(figsize=(20,10))
    ax.xaxis.set_tick_params(labelsize=18, rotation=25)
    ax.yaxis.set_tick_params(labelsize=18)
    range_bar1 = np.arange(len(classes))
    rects1 = ax.bar(range_bar1, tuple(scores[:, 0]), width, color='b')
    rects2 = ax.bar(range_bar1 + width, tuple(scores[:, 1]), width, color='r')
    ax.set_ylabel('Scores',fontsize=22)
    ax.set_title('Tag scores', fontsize=22)
    ax.set_xticks(range_bar1 + width / 2)
    ax.set_xticklabels(classes)
    ax.legend((rects1[0], rects2[0]), ('Correct', 'Wrong'), fontsize=20)
    plt.legend()
    plt.savefig('/kaggle/working/tag_scores.png', bbox_inches="tight", transparent=True)

raw_data = pd.read_csv('/kaggle/input/lsa64handskeletaldata/final_dataset_LSA64.csv', header=0)
dataset = raw_data.values

# Đặc trưng (features)
X = dataset[0:340187, 1:127].astype(float)
# "Time" cho mô hình học
Y = dataset[0:340187, 127]

encoder = LabelEncoder()
encoder_Y = encoder.fit(Y)
encoded_Y = encoder_Y.transform(Y)
dummy_Y = to_categorical(encoded_Y)

# Chia dữ liệu thành 80% huấn luyện và 20% kiểm tra + xác thực
X_train_val, X_test, Y_train_val, Y_test = train_test_split(X, dummy_Y, test_size=0.2, random_state=9)

# Chia dữ liệu huấn luyện và xác thực thành 87.5% huấn luyện và 12.5% xác thực (tương ứng với 10% tổng dữ liệu)
X_train, X_val, Y_train, Y_val = train_test_split(X_train_val, Y_train_val, test_size=0.125, random_state=9)

# Thêm chiều vào dữ liệu để phù hợp với mô hình LSTM
X_train = np.expand_dims(X_train, axis=1)
X_test = np.expand_dims(X_test, axis=1)
X_val = np.expand_dims(X_val, axis=1)

# Xây dựng mô hình Keras
model = Sequential()
model.add(LSTM(units=512, input_shape=(None, 126), return_sequences=True))  
model.add(BatchNormalization()) 
model.add(LSTM(units=256, return_sequences=True))
model.add(Dropout(0.5))  
model.add(LSTM(units=128, return_sequences=True))
model.add(Dropout(0.5))  
model.add(LSTM(units=64))
model.add(Dense(units=64, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(units=64, activation='softmax'))

his = LossHistory()
model.compile(loss='categorical_crossentropy', optimizer=Adam(0.0001), metrics=['accuracy'])
model.fit(X_train, Y_train, batch_size=32, epochs=60, verbose=1, validation_data=(X_val, Y_val), callbacks=[his])
model.summary()
his.loss_plot('epoch')
model.save('/kaggle/working/framewise_recognition_tmp.h5')

# Đánh giá và vẽ ma trận nhầm lẫn
print('Test:')
score, accuracy = model.evaluate(X_test, Y_test, batch_size=32)
print('Test Score:{:.3}'.format(score))
print('Test accuracy:{:.3}'.format(accuracy))

# Ma trận nhầm lẫn
Y_pred = model.predict(X_test)
y_true = np.argmax(Y_test, axis=1)
y_pred = np.argmax(Y_pred, axis=1)

precision = precision_score(y_true, y_pred, average='macro')
recall = recall_score(y_true, y_pred, average='macro')
f1 = f1_score(y_true, y_pred, average='macro')
print(f'Precision: {precision:.6f}')
print(f'Recall: {recall:.6f}')
print(f'F1 Score: {f1:.6f}')
cfm = confusion_matrix(np.argmax(Y_test, axis=1), np.argmax(Y_pred, axis=1))
np.set_printoptions(precision=2)
plt.figure()
class_names = ['Opaque', 'Red', 'Green', 'Yellow', 'Bright', 'Light_blue', 'Colors', 
               'Red2', 'Women', 'Enemy', 'Son', 'Man', 'Away', 'Drawer', 'Born', 'Learn', 
               'Call', 'Skimmer', 'Bitter', 'Sweet_milk', 'Milk', 'Water', 'Food', 
               'Argentina', 'Urugauy', 'Country', 'Last_name', 'Where', 'Mock', 'Birthday', 
               'Breakfast', 'Photo', 'Hungry', 'Map', 'Coin', 'Music', 'Ship', 'none', 
               'Name', 'Patience', 'Perfume', 'Deaf', 'Trap', 'Rice', 'Barbecue', 'Candy', 
               'Chewing_gum', 'Spaghetti', 'Yogurt', 'Accept', 'Thanks', 'Shut_down', 'Appear', 
               'To_land', 'Catch', 'Help', 'Dance', 'Bathe', 'Buy', 'Copy', 'Run', 'Realize', 
               'Give', 'Find']
cfm = confusion_matrix(np.argmax(Y_test, axis=1), np.argmax(Y_pred, axis=1))
plot_confusion_matrix(Actions.__members__.keys(), cfm)
his.loss_plot('epoch')

# Báo cáo phân loại
y_pred = np.argmax(Y_pred, axis=1)
y_true = np.argmax(Y_test, axis=1)
classes = list(Actions.__members__.keys())
class_names, plotMat, support = get_report(y_true, y_pred, classes)

# Lưu báo cáo phân loại
report_df = pd.DataFrame(class_names, columns=["Class"])
report_df["Precision"] = plotMat[:, 0]
report_df["Recall"] = plotMat[:, 1]
report_df["F1-score"] = plotMat[:, 2]
report_df["Support"] = support
report_df.to_csv('/kaggle/working/classification_report.csv', index=False)

# Lưu biểu đồ báo cáo phân loại
plt.figure(figsize=(10, 8))
plt.barh(class_names, plotMat[:, 2], color='skyblue', label='F1-score')
plt.barh(class_names, plotMat[:, 1], color='lightcoral', alpha=0.7, label='Recall')
plt.barh(class_names, plotMat[:, 0], color='lightgreen', alpha=0.5, label='Precision')
plt.xlabel('Score')
plt.ylabel('Classes')
plt.title('Classification Report')
plt.legend()
plt.savefig('/kaggle/working/classification_report.png', bbox_inches="tight", transparent=True)
plt.show()
```

```python
import pandas as pd
from enum import Enum
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense, Dropout, BatchNormalization, LSTM
from keras.optimizers import Adam
from keras.callbacks import Callback
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, f1_score, precision_score, recall_score

class Actions(Enum):
    Opaque = 0 
    Red = 1 
    Green = 2 
    Yellow = 3 
    Bright = 4 
    Light_blue = 5 
    Colors = 6 
    Red2 = 7
    Women = 8 
    Enemy = 9 
    Son = 10 
    Man = 11 
    Away = 12 
    Drawer = 13 
    Born = 14 
    Learn = 15 
    Call = 16 
    Skimmer = 17 
    Bitter = 18 
    Sweet_milk = 19 
    Milk = 20 
    Water = 21 
    Food = 22 
    Argentina = 23 
    Urugauy = 24
    Country = 25 
    Last_name = 26 
    Where = 27 
    Mock = 28 
    Birthday = 29 
    Breakfast = 30 
    Photo = 31 
    Hungry = 32 
    Map = 33 
    Coin = 34 
    Music = 35 
    Ship = 36 
    none = 37 
    Name = 38 
    Patience = 39 
    Perfume = 40 
    Deaf = 41 
    Trap = 42 
    Rice = 43 
    Barbecue = 44 
    Candy = 45 
    Chewing_gum = 46 
    Spaghetti = 47 
    Yogurt = 48 
    Accept = 49 
    Thanks = 50 
    Shut_down = 51 
    Appear = 52 
    To_land = 53 
    Catch = 54 
    Help = 55 
    Dance = 56 
    Bathe = 57 
    Buy = 58 
    Copy = 59 
    Run = 60 
    Realize = 61 
    Give = 62 
    Find = 63 

class LossHistory(Callback):
    def on_train_begin(self, logs={}):
        self.losses = {'batch':[], 'epoch':[]}
        self.accuracy = {'batch':[], 'epoch':[]}
        self.val_loss = {'batch':[], 'epoch':[]}
        self.val_acc = {'batch':[], 'epoch':[]}

    def on_batch_end(self, batch, logs={}):
        self.losses['batch'].append(logs.get('loss'))
        self.accuracy['batch'].append(logs.get('acc'))
        self.val_loss['batch'].append(logs.get('val_loss'))
        self.val_acc['batch'].append(logs.get('val_acc'))

    def on_epoch_end(self, batch, logs={}):
        self.losses['epoch'].append(logs.get('loss'))
        self.accuracy['epoch'].append(logs.get('acc'))
        self.val_loss['epoch'].append(logs.get('val_loss'))
        self.val_acc['epoch'].append(logs.get('val_acc'))

    def loss_plot(self, loss_type):
        plt.figure(figsize=(12, 8))  # Điều chỉnh kích thước của hình ảnh
        plt.style.use('ggplot')  # Sử dụng ggplot style với nền màu trắng
        iters = range(len(self.losses[loss_type]))
        plt.plot(iters, self.accuracy[loss_type], 'r', label='train acc')
        plt.plot(iters, self.losses[loss_type], 'g', label='train loss')
        if loss_type == 'epoch':
            plt.plot(iters, self.val_acc[loss_type], 'b', label='val acc')
            plt.plot(iters, self.val_loss[loss_type], 'k', label='val loss')
        plt.grid(True)
        plt.xlabel(loss_type, fontsize=14)  # Điều chỉnh kích thước của nhãn trục x
        plt.ylabel('acc-loss', fontsize=14)  # Điều chỉnh kích thước của nhãn trục y
        plt.legend(loc="upper right", fontsize=12)
        plt.savefig('/kaggle/working/loss_accuracy_plot.png', bbox_inches="tight", transparent=True)

def get_report(y_true, y_pred, classes):
    clf_report = classification_report(y_true, y_pred, labels=classes, zero_division=0)
    clf_report = clf_report.replace('\n\n', '\n')
    clf_report = clf_report.replace('micro avg', 'micro_avg')
    clf_report = clf_report.replace('macro avg', 'macro_avg')
    clf_report = clf_report.replace('weighted avg', 'weighted_avg')
    clf_report = clf_report.replace(' / ', '/')
    lines = clf_report.split('\n')

    class_names, plotMat, support = [], [], []
    for line in lines[1:]:
        t = line.strip().split()
        if len(t) < 2:
            continue
        v = [float(x) for x in t[1: len(t) - 1]]
        if len(v) == 1 : v = v * 3
        support.append(int(t[-1]))
        class_names.append(t[0])
        plotMat.append(v)
    plotMat = np.array(plotMat)
    support = np.array(support)
    return class_names, plotMat, support

def plot_confusion_matrix(classes, mat, normalize=True, cmap=plt.cm.Blues):
    cm = np.copy(mat)
    title = 'Ma trận nhầm lẫn (không chuẩn hóa)'
    if normalize: 
        cm = cm.astype('float') / np.sum(cm, axis=1, keepdims=True)
        title = title.replace('không', 'có')
    plt.clf()
    fig, ax = plt.subplots(figsize=(30, 25))  # Điều chỉnh kích thước hình ảnh
    ax.set_title(title, y=-0.06, fontsize=22)
    ax.xaxis.set_ticks_position('top')
    ax.xaxis.set_label_position('top')
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.clim(vmin=0.0, vmax=1.0)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=90, fontsize=12)  # Điều chỉnh góc và kích thước của nhãn
    plt.yticks(tick_marks, classes, fontsize=12)
    fmt = '.2f' if normalize else 'd'
    thresh = np.max(cm) / 2
    thresh = 1 / 2
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            color = "white" if (cm[i, j] > thresh) else "black"
            plt.text(j, i, format(cm[i, j], fmt), horizontalalignment="center", color=color, fontsize=10)  # Điều chỉnh kích thước của con số
    plt.ylabel('Nhãn thực', fontsize=16)  # Điều chỉnh kích thước của nhãn trục y
    plt.xlabel('Nhãn dự đoán', fontsize=16)  # Điều chỉnh kích thước của nhãn trục x
    plt.tight_layout()
    plt.savefig('/kaggle/working/confusion_matrix.png', bbox_inches="tight", transparent=True)

def get_scores(y_true, y_pred, classes):
    correct, wrong = {}, {}
    for tag in classes:
        correct[tag] = 0
        wrong[tag] = 0

    for tag, pred in zip(y_true, y_pred):
        if tag == pred:
            correct[tag] += 1
        else:
            wrong[tag] += 1

    scores = []
    total = len(y_true)
    for tag in classes:
        cur = np.array([correct[tag], wrong[tag]])
        scores.append(cur / total)
    return np.array(scores)

def plot_tag_scores(classes, scores, normalize=True):
    plt.clf()
    width = 0.45
    fig, ax = plt.subplots(figsize=(20,10))
    ax.xaxis.set_tick_params(labelsize=18, rotation=25)
    ax.yaxis.set_tick_params(labelsize=18)
    range_bar1 = np.arange(len(classes))
    rects1 = ax.bar(range_bar1, tuple(scores[:, 0]), width, color='b')
    rects2 = ax.bar(range_bar1 + width, tuple(scores[:, 1]), width, color='r')
    ax.set_ylabel('Scores',fontsize=22)
    ax.set_title('Tag scores', fontsize=22)
    ax.set_xticks(range_bar1 + width / 2)
    ax.set_xticklabels(classes)
    ax.legend((rects1[0], rects2[0]), ('Correct', 'Wrong'), fontsize=20)
    plt.legend()
    plt.savefig('/kaggle/working/tag_scores.png', bbox_inches="tight", transparent=True)

raw_data = pd.read_csv('/kaggle/input/lsa64handskeletaldata/final_dataset_LSA64.csv', header=0)
dataset = raw_data.values

# Đặc trưng (features)
X = dataset[0:340187, 1:127].astype(float)
# "Time" cho mô hình học
Y = dataset[0:340187, 127]

encoder = LabelEncoder()
encoder_Y = encoder.fit(Y)
encoded_Y = encoder_Y.transform(Y)
dummy_Y = to_categorical(encoded_Y)

# Chia dữ liệu thành 80% huấn luyện và 20% kiểm tra + xác thực
X_train_val, X_test, Y_train_val, Y_test = train_test_split(X, dummy_Y, test_size=0.2, random_state=9)

# Chia dữ liệu huấn luyện và xác thực thành 87.5% huấn luyện và 12.5% xác thực (tương ứng với 10% tổng dữ liệu)
X_train, X_val, Y_train, Y_val = train_test_split(X_train_val, Y_train_val, test_size=0.125, random_state=9)

# Thêm chiều vào dữ liệu để phù hợp với mô hình LSTM
X_train = np.expand_dims(X_train, axis=1)
X_test = np.expand_dims(X_test, axis=1)
X_val = np.expand_dims(X_val, axis=1)

# Xây dựng mô hình Keras
model.add(Dense(units=256, activation='relu'))
model.add(BatchNormalization()) 
model.add(Dense(units=128, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))  
model.add(Dense(units=64, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(units=64, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(units=64, activation='softmax'))

his = LossHistory()
model.compile(loss='categorical_crossentropy', optimizer=Adam(0.0001), metrics=['accuracy'])
model.fit(X_train, Y_train, batch_size=32, epochs=60, verbose=1, validation_data=(X_val, Y_val), callbacks=[his])
model.summary()
his.loss_plot('epoch')
model.save('/kaggle/working/framewise_recognition_tmp.h5')

# Đánh giá và vẽ ma trận nhầm lẫn
print('Test:')
score, accuracy = model.evaluate(X_test, Y_test, batch_size=32)
print('Test Score:{:.3}'.format(score))
print('Test accuracy:{:.3}'.format(accuracy))

# Ma trận nhầm lẫn
Y_pred = model.predict(X_test)
y_true = np.argmax(Y_test, axis=1)
y_pred = np.argmax(Y_pred, axis=1)

precision = precision_score(y_true, y_pred, average='macro')
recall = recall_score(y_true, y_pred, average='macro')
f1 = f1_score(y_true, y_pred, average='macro')
print(f'Precision: {precision:.6f}')
print(f'Recall: {recall:.6f}')
print(f'F1 Score: {f1:.6f}')
cfm = confusion_matrix(np.argmax(Y_test, axis=1), np.argmax(Y_pred, axis=1))
np.set_printoptions(precision=2)
plt.figure()
class_names = ['Opaque', 'Red', 'Green', 'Yellow', 'Bright', 'Light_blue', 'Colors', 
               'Red2', 'Women', 'Enemy', 'Son', 'Man', 'Away', 'Drawer', 'Born', 'Learn', 
               'Call', 'Skimmer', 'Bitter', 'Sweet_milk', 'Milk', 'Water', 'Food', 
               'Argentina', 'Urugauy', 'Country', 'Last_name', 'Where', 'Mock', 'Birthday', 
               'Breakfast', 'Photo', 'Hungry', 'Map', 'Coin', 'Music', 'Ship', 'none', 
               'Name', 'Patience', 'Perfume', 'Deaf', 'Trap', 'Rice', 'Barbecue', 'Candy', 
               'Chewing_gum', 'Spaghetti', 'Yogurt', 'Accept', 'Thanks', 'Shut_down', 'Appear', 
               'To_land', 'Catch', 'Help', 'Dance', 'Bathe', 'Buy', 'Copy', 'Run', 'Realize', 
               'Give', 'Find']
cfm = confusion_matrix(np.argmax(Y_test, axis=1), np.argmax(Y_pred, axis=1))
plot_confusion_matrix(Actions.__members__.keys(), cfm)
his.loss_plot('epoch')

# Báo cáo phân loại
y_pred = np.argmax(Y_pred, axis=1)
y_true = np.argmax(Y_test, axis=1)
classes = list(Actions.__members__.keys())
class_names, plotMat, support = get_report(y_true, y_pred, classes)

# Lưu báo cáo phân loại
report_df = pd.DataFrame(class_names, columns=["Class"])
report_df["Precision"] = plotMat[:, 0]
report_df["Recall"] = plotMat[:, 1]
report_df["F1-score"] = plotMat[:, 2]
report_df["Support"] = support
report_df.to_csv('/kaggle/working/classification_report.csv', index=False)

# Lưu biểu đồ báo cáo phân loại
plt.figure(figsize=(10, 8))
plt.barh(class_names, plotMat[:, 2], color='skyblue', label='F1-score')
plt.barh(class_names, plotMat[:, 1], color='lightcoral', alpha=0.7, label='Recall')
plt.barh(class_names, plotMat[:, 0], color='lightgreen', alpha=0.5, label='Precision')
plt.xlabel('Score')
plt.ylabel('Classes')
plt.title('Classification Report')
plt.legend()
plt.savefig('/kaggle/working/classification_report.png', bbox_inches="tight", transparent=True)
plt.show()
```
